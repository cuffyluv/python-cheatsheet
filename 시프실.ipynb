```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정 (예: 나눔고딕)
plt.rc('font', family='NanumGothic')  # 폰트 변경
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# 폰트 확인
font_list = [f.name for f in fm.fontManager.ttflist]
print(font_list)  # 시스템 내 설치된 폰트 확인
import matplotlib.pyplot as plt

# 데이터 설정
labels = ['로컬 프로그램', '파이 1개 단일', '파이 1개 10개', '파이 2개 10개', '파이 3개 10개']
times = [16, 54, 23, 19, 16]

# 그래프 그리기
plt.figure(figsize=(10, 6))
bars = plt.bar(labels, times, color=['gray', 'red', 'blue', 'green', 'purple'])

# 값 표시
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval}초', ha='center', va='bottom')

# 제목 및 축 레이블
plt.title('분산 처리 - 연산 속도 비교', fontsize=16)
plt.xlabel('실험 조건')
plt.ylabel('연산 시간 (초)')

# 그래프 출력
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
